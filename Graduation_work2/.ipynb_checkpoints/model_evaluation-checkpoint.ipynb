```python
#ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
import seaborn as sns
import os
import pathlib
import cv2

sns.set()
%matplotlib inline

from matplotlib import rcParams
rcParams['font.family'] = 'MS Gothic'

import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'

from pathlib import Path
```

```python
#モデルのロード
model = tf.keras.models.load_model("video_model.h5")
```

```python
#test_x, test_yを生み出すコード
#csvファイルの確認
df = pd.read_csv("gray_emotional_sign/感情手話データセット.csv", encoding = "utf-8")
print(df.head())

#動画ファイルが存在するか確認
p = Path("./gray_emotional_sign")
#print(p)
print(p.is_dir())

p_list = [str(path) for path in p.rglob("*.mp4")]
p_array = np.array(p_list)
test_x = np.concatenate([p_array[35:40], p_array[75:80], p_array[115:120], p_array[155:160], p_array[195:200]])

test_y = np.concatenate([
    df["ラベル"][35:40].to_numpy(),
    df["ラベル"][75:80].to_numpy(),
    df["ラベル"][115:120].to_numpy(),
    df["ラベル"][155:160].to_numpy(),
    df["ラベル"][195:200].to_numpy()
])
print("test_x :", test_x[:5])
print(test_x.shape)
print("test_y :", test_y[:5])
print(test_y.shape)

def video_resize(video_name, target_frames = 305):
    #動画ファイルをオープン
    cap = cv2.VideoCapture(video_name)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    interval = max(total_frames // target_frames, 1)
    frames = []
    count = 0

    while cap.isOpened() and len(frames) < target_frames:
        ret, frame = cap.read()
        if ret == False:
          break
        if count % interval == 0:  # 一定間隔でフレームを抽出
          #リサイズ
          resized_cap = cv2.resize(frame, (64, 64))
          #グレースケール化
          gray_cap = cv2.cvtColor(resized_cap, cv2.COLOR_BGR2GRAY)
          #正規化
          Normalized_cap = gray_cap[:, :, np.newaxis] / 255.0
          frames.append(Normalized_cap)
        count += 1
    #動画ファイルをクローズ
    cap.release()

    #フレームが不足している場合、ゼロ埋めで補完
    #最大値(305.0フレーム)
    while len(frames) < target_frames:
        frames.append(np.zeros((64, 64, 1)))

    return frames

test_resized = list(video_resize(video_name) for video_name in test_x)
test_resized = np.array(test_resized, dtype = np.float32)
print("テスト用動画データの形状:", test_resized.shape)
test_x = test_resized


print("テストデータの最小値:", test_resized.min())
print("テストデータの最大値:", test_resized.max())

print(type(test_x))
print(type(test_y))


```

```python
#モデルの評価
test_loss, test_acc = model.evaluate(test_x, test_y)
print(f"テストデータでの損失: {test_loss}, 正解率: {test_acc}")
```
